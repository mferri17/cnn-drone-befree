```python
import numpy as np

nums = np.random.normal(loc=0, scale=1, size=1000000)
print('Gaussian:')
print('min: \t', np.min(nums))
print('mean: \t', np.mean(nums))
print('max: \t', np.max(nums))

nums = np.random.triangular(left=0, mode=0.5, right=1, size=1000000)
print('Triangular:')
print('min: \t', np.min(nums))
print('mean: \t', np.mean(nums))
print('max: \t', np.max(nums))
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

np.random.seed(42)

size = 6
true = np.arange(size)
pred = np.arange(size) + np.random.normal(0, 0.1, size)

# true = np.flip(true)
# pred = np.flip(pred)

print('true', true)
print('pred', pred)

print('r2', r2_score(true, pred))
print('mse', mean_squared_error(true, pred))
print('mae', mean_absolute_error(true, pred))
```
